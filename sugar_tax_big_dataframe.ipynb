```python
def strip_character_from_dataframe(df, char_to_strip):
    return df.applymap(lambda x: x.replace(char_to_strip, '') if isinstance(x, str) else x)

# Call the function to remove the character 'a'
```

```python
def strip_chars(df):
  df = strip_character_from_dataframe(df, '%')
  df = strip_character_from_dataframe(df, '*')
  df = strip_character_from_dataframe(df, ',')
  df = df.replace('-', 0)
  df.applymap(lambda x: x.replace('-', '0') if isinstance(x, str) else x)
  return df
```

```python
obese = strip_chars(obese)
increased_risk = strip_chars(increased_risk)
```

```python
obese = obese.astype(float)
increased_risk = increased_risk.astype(float)
obese
```

```python
obese = obese.reset_index(drop=True)

increased_risk = increased_risk.reset_index(drop = True)
```

```python
ob_ir = obese+increased_risk
ob_ir
```

```python
# Function to reshape the DataFrame into a 8x2 DataFrame
def reshape_dataframe(df):
    # Create an empty list to hold the new rows
    reshaped_rows = []

    # Iterate over every two columns at a time
    for i in range(0, len(df.columns), 2):
        # Take the next two columns
        cols_to_add = df.iloc[:, i:i+2]

        # Append each row of these columns to the reshaped rows list
        reshaped_rows.extend(cols_to_add.values.tolist())

    # Create a new DataFrame from the reshaped rows
    reshaped_df = pd.DataFrame(reshaped_rows)

    return reshaped_df
```

```python
increased_risk = reshape_dataframe(increased_risk)
obese = reshape_dataframe(obese)


increased_risk.columns = ['Percentage of Increased Risk Population (23.0-27.49 BMI)', 'Total Increased Risk Population (23.0-27.49 BMI)']
obese.columns = ['Percentage of Obese Population (BMI 27.5 or Higher)', 'Total Obese Population (BMI 27.5 or Higher)']
```

```python
ob_ir = reshape_dataframe(ob_ir)
ob_ir.columns = ['Percentage of Obese Population (23.0 BMI or Higher)', 'Total Obese Population (23.0 BMI or Higher)']
ob_ir
```

```python
df = pd.concat([df, ob_ir], axis=1)
obese_pop_csv = df
df
```
